```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt


def ode_2(gamma, y, w):
    y1, y2 = y
    return [y2, np.sin(gamma) - w ** 2 * np.sin(y1)]

span = [0, 20]
y0 = [0, 0]
w = 6

sol = solve_ivp(lambda gamma, y: ode_2(gamma, y, w), span, y0)
print(sol)


gamma_above_w = np.linspace(w, span[1], 1000)
gamma_below_w = np.linspace(w, span[0], 1000)
gamma_equals_w = np.array([w])

fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(sol.t, sol.y[0], label='Решение')
ax.plot(gamma_above_w, sol.sol(gamma_above_w), label='gamma < w')
ax.plot(gamma_below_w, sol.sol(gamma_below_w)[0], label='gamma > w')
ax.plot(gamma_equals_w, sol.sol(gamma_equals_w)[0], label='gamma = w')


ax.legend()
ax.set_xlabel('gamma')
ax.set_ylabel('w')
plt.show()
```
